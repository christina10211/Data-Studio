## Scraping the first page of the patents page.

```python
from time import sleep
from random import randint
```

```python
url = "https://patents.justia.com/company/facebook?list=patents&page=1"

response = requests.get(url).content

soup = BeautifulSoup(response, 'html5lib')
```

```python
blocks = soup.find_all('li',attrs={'class':'has-padding-content-block-30 -zb'})
blocks[0]
```

```python
#There are 47 pages in total, so we will make 47 requests.
#we will use time.sleep to control the crawling speed

patent_list = []
for i in range(1,48):
    
    url = "https://patents.justia.com/company/facebook?list=patents&page="+str(i)

    response = requests.get(url).content

    soup = BeautifulSoup(response, 'html5lib')
    
    sleep(randint(1,5))

    blocks = soup.find_all('li',attrs={'class':'has-padding-content-block-30 -zb'})
    blocks[0]
    
    for block in blocks:
        patent_dic = {}

        #title
        title = block.find(class_='head').text.strip()
        patent_dic['title'] = title

        #url
        url = 'https://patents.justia.com'+block.find('a')['href']
        patent_dic['url']=url

        #patent number

        p_nums = block.find(class_ = 'number').text.strip()
        p_num = re.findall(r"number:\s+(.*)$",p_nums)[0]
        patent_dic['p_num'] = p_num

        #abstract
        try:
            abstracts = block.find(class_ = 'abstract').text.strip()
            abstract = re.findall('Abstract:(.*)',abstracts)[0].strip()
            patent_dic['abstract'] = abstract
        except:
            patent_dic['abstract'] = 'None'

        #inventors
        inventors = block.find(class_ = 'inventors').text.strip()
        inventor = re.findall(":\s+(\w.*)$",inventors)[0]
        patent_dic['inventor'] = inventor


        #date filed
        dates_1 = block.find(class_ = 'date-filed').text.strip()
        date_filed = re.findall('Filed:\s+(.*)$',dates_1)[0].strip()
        patent_dic['date_filed'] = date_filed


        #date of patent (date patent is granted)
        dates_2 = block.find(class_ = 'date-issued').text.strip()
        date_issued = re.findall('Patent:\s+(.*)$',dates_2)[0].strip()
        patent_dic['date_issued'] = date_issued


        #Assignee
        assignees = block.find(class_ = 'assignees').text.strip()
        assignee = re.findall('Assignee:\s+(.*)$',assignees)[0].strip().upper()
        patent_dic['assignee'] = assignee

        #type
        types = block.find(class_ = 'type').text.strip()
        p_type = re.findall('Type:\s+(.*)$',types)[0].strip()
        patent_dic['type'] = p_type

        patent_list.append(patent_dic)
    
    
    
    
    

    
    
    
    
    
    
    
   

   
    
    
```

```python
df_patent = pd.DataFrame(patent_list)
```

```python
#There are 15 pages in total, so we will make 47 requests.
#we will use time.sleep to control the crawling speed

app_list = []
for i in range(1,16):
    
    url = "https://patents.justia.com/company/facebook?list=applications&page="+str(i)

    response = requests.get(url).content

    soup = BeautifulSoup(response, 'html5lib')
    
    sleep(randint(1,5))

    blocks = soup.find_all('li',attrs={'class':'has-padding-content-block-30 -zb'})
    blocks[0]
    
    for block in blocks:
        app_dic = {}

        #title
        title = block.find(class_='head').text.strip()
        app_dic['title'] = title

        #url
        url = 'https://patents.justia.com'+block.find('a')['href']
        app_dic['url']=url

        #patent number

        p_nums = block.find(class_ = 'number').text.strip()
        p_num = re.findall(r"number:\s+(.*)$",p_nums)[0]
        app_dic['p_num'] = p_num

        #abstract
        try:
            abstracts = block.find(class_ = 'abstract').text.strip()
            abstract = re.findall('Abstract:(.*)',abstracts)[0].strip()
            app_dic['abstract'] = abstract
        except:
            app_dic['abstract'] = 'None'

        #inventors
        inventors = block.find(class_ = 'inventors').text.strip()
        inventor = re.findall(":\s+(\w.*)$",inventors)[0]
        app_dic['inventor'] = inventor


        #date filed
        dates_1 = block.find(class_ = 'date-filed').text.strip()
        date_filed = re.findall('Filed:\s+(.*)$',dates_1)[0].strip()
        app_dic['date_filed'] = date_filed

        #Applicants
        applicants = block.find(class_ = 'assignees').text.strip()
        applicant = re.findall('Applicant:\s+(.*)$',applicants)[0].strip().upper()
        app_dic['applicant'] = applicant
        
        #type
        types = block.find(class_ = 'type').text.strip()
        p_type = re.findall('Type:\s+(.*)$',types)[0].strip()
        app_dic['type'] = p_type
        

        app_list.append(app_dic)
```

```python
df_app = pd.DataFrame(app_list)
```

```python
def get_classification(n):
    url = 'https://patents.google.com/patent/US'+str(n)
    
    response = requests.get(url).content
    sleep(randint(1,5))
    soup = BeautifulSoup(response, 'html5lib')
    # print(soup.find_all('span'))
    blocks = soup.find_all('span',{'itemprop':'Description'})
    list_=[] 
    for block in blocks:
        classification = (block.text.strip())
        if classification not in list_:
            list_.append(classification)
    return list_

    
    
#     sleep(randint(1,5))

#     class_ = block.find(class_='wrap').text.strip()
 
```

```python
sub = []
for index, row in df.iterrows():
    if len(row['classifications'])!=0:
        sub.append(row["classifications"][1])
```

```python
third = []
for index, row in df.iterrows():
    if len(row['classifications'])!=0:
        third.append(row["classifications"][2])
```

```python
year = []
types = []
titles = []
abstracts = []
for index, row in df.iterrows():
    if len(row['classifications'])!=0:
        year.append(row["year_filed"])
        types.append(row['type'])
        titles.append(row['title'])
        abstracts.append(row['abstract'])
        

```
